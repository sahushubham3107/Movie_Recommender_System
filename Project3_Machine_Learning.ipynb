The genre developed from above code is not in the proper format so we need to display it in correct format .

but here we need to convert in proper list.

Importing ast which convert strings into list


```python
import ast
def convert(obj):  # in this line obj will be a string
  L=[]
  for i in ast.literal_eval(obj):   # in this line obj will become a list and we can perform loop in it , ast will perform this job
    L.append(i['name'])
  return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)  # to see genres as a list
movies.head()
```

```python
# Same thing we are doing with keywords to convert it into a list
movies['keywords'] = movies['keywords'].apply(convert)  
```

For the cast column we want to see only first 3 cast actors so let's print the name from the cast for  movies

```python
import ast
def convert3(obj):  # in this line obj will be a string
  counter=0
  L=[]
  for i in ast.literal_eval(obj):   # in this line obj will become a list and we can perform loop in it , ast will perform this job
    if  (counter != 3):
      L.append(i['name'])
      counter+=1
    else:
      break
    
  return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)  # this will print the first 3 cast stars of any movie in the same column
movies.head()
```

We are now making functions for fetching the name of director from the crew array

```python
def fetch_director(obj):  # in this line obj will be a string
  L=[]
  for i in ast.literal_eval(obj):   # in this line obj will become a list and we can perform loop in it , ast will perform this job
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

**If we want to get top 5000 words which are repeated**

Now program will consider two similar words as two different words 
- For example - it is considering action and actions as two different words 
- Here we need to do the steming 

```python
# Doing the steming count the similar type of words only single time
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

We will now verify with first array of tags that is it working or not 

```python
new_df['tags'][0]
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

The above code is working well so now we will form a code to apply it in the whole dataset

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

**Now we are going to do text vectorization**
- Here we are going to use bag of words technique . There are other techniques as well like tfidf

In sklearn library there is a function named as feature_extraction under which a function exist named CountVectorizer. Lets use it for our vectorization purpose. Also it won't count some common english words which are known as stop words.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

Now fetching top 5000 words which are repeatedly used

```python
cv.get_feature_names()
```

**Measuring distance between the features**

Either we can calculate the distance between two vectors which is also known as eucledian distance but cosine distance(measuring angle) is more reliable feature for large dataset.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
# you can check the similarity of movie 1 with all other movies as
similarity[0]
```

```python
# but when sorting occurs according to the distance it will looses its index so now we will use the function enumerate
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

Defining a function which recommends five similar movies from the given input movie

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]  # print the top five recommended movies

  for i in movies_list:   
    print(new_df.iloc[i[0]].title)   # this will print all the recommended movies
    
```

```python
# this will return the five recommended movies by just entering the movie name
recommend('Star Wars')
```
